```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
```

```python
reasons = ['app', 'crash', 'slow', 'inaction', 'response', 'worst']
# reasons = [0,1,2,3,4,5]
```

```python
for each in reasons:
    G.add_node(each)
    
for i in range(len(reasons)):
    if i > 0:
        G.add_edge(reasons[0], reasons[i], length = 2)
```

```python
nx.draw(G, node_size=None, with_labels = True)
plt.show()
```
